```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/Excelerate_IITChicago_Internship/Marketing Team Data.xlsx")
df.head(3)
```

```python
# Aggregate the data by summing the 'Impressions' for each unique combination of 'Campaign ID' and 'Geography'
aggregated_data = df.groupby(['Campaign ID', 'Geography'])['Impressions'].sum().reset_index()

# Pivot the aggregated data to have geographies as columns
stacked_pivot = aggregated_data.pivot(index='Campaign ID', columns='Geography', values='Impressions')

# Sort the geographies alphabetically
stacked_pivot = stacked_pivot.reindex(sorted(stacked_pivot.columns), axis=1)

# Create a stacked bar chart
ax = stacked_pivot.plot(kind='bar', stacked=True, figsize=(10, 6))
ax.set_xticklabels(sorted(stacked_pivot.index, key=lambda x: int(x.split(' ')[1])), rotation=45)
plt.xlabel('Campaign ID')
plt.ylabel('Impressions')
plt.title('Impressions by Geography')
plt.tight_layout()

# Move the legend outside the plot area
ax.legend(bbox_to_anchor=(1, 1), title='Geography')
plt.show();
```

```python
# Aggregate the data by summing the 'Impressions' for each unique combination of 'Campaign ID' and 'Geography'
aggregated_data = df.groupby(['Campaign ID', 'Geography'])['Reach'].sum().reset_index()

# Pivot the aggregated data to have geographies as columns
stacked_pivot = aggregated_data.pivot(index='Campaign ID', columns='Geography', values='Reach')

# Sort the geographies alphabetically
stacked_pivot = stacked_pivot.reindex(sorted(stacked_pivot.columns), axis=1)

# Create a stacked bar chart
ax = stacked_pivot.plot(kind='bar', stacked=True, figsize=(10, 6))
ax.set_xticklabels(sorted(stacked_pivot.index, key=lambda x: int(x.split(' ')[1])), rotation=45)
plt.xlabel('Campaign ID')
plt.ylabel('Reach')
plt.title('Reach by Geography')
plt.tight_layout()

# Move the legend outside the plot area
ax.legend(bbox_to_anchor=(1, 1), title='Geography')
plt.show();
```
